```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load 
schools = "Resources/schools_complete.csv"
students = "Resources/students_complete.csv"


# Read School and Student Data File and store into Pandas Data Frames
school_df = pd.read_csv(schools)
student_df = pd.read_csv(students)

# Combine the data into a single dataset
school_data_all = pd.merge(student_df, school_df, how = 'left', on = 'school_name')
#school_data_all.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#calulate district summary data

num_of_schools = len(school_data_all["school_name"].unique())

num_of_students = school_data_all["Student ID"].value_counts().sum()

total_budget = school_df["budget"].sum()

avg_math_all = student_df["math_score"].mean()

avg_read_all = student_df["reading_score"].mean()

passing_math = student_df.loc[student_df["math_score"] >= 70]

passing_reading = student_df.loc[student_df["reading_score"] >= 70]

percent_passing_math = passing_math["student_name"].count()/num_of_students

percent_passing_reading = passing_reading["student_name"].count()/num_of_students

percent_passing_overall = (percent_passing_math + percent_passing_reading)/2
```

```python
district_data = pd.DataFrame(data = {"Total Schools" : [num_of_schools], "Total Students" : [num_of_students], 
                                     "Total Budget" : [total_budget], "Average Math Score" :[avg_math_all], 
                                     "Average Reading Score" : [avg_read_all], "% Passing Math" : [percent_passing_math], 
                                     "% Passing Readng" :[percent_passing_reading], " Overall Passing Rate" :[percent_passing_overall]})
district_data
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
school_data = school_df.drop(columns = "School ID") # remove uneeded column

budget_per_student = school_data["budget"]/school_data["size"] #create new column called budget per students
school_data["Per Student Budget"] = budget_per_student # add budget per student column to dataframe

# regroup "all school data" based on school and calculate averages of specified columns

by_school = school_data_all.groupby("School ID") 

avg_math = by_school["math_score"].mean()
avg_read = by_school["reading_score"].mean()

# add calculated averages to our dataframe
school_data["Average Math Score"] = avg_math
school_data["Average Reading Score"] = avg_read


# create new dataframe to only include passing students
passing_math = school_data_all.loc[school_data_all["math_score"] >= 70]
passing_reading = school_data_all.loc[school_data_all["reading_score"] >= 70]

# count number of students in this new dataframe then divide by the total number of students
# at the school (school_data[size])...creating a new column. Do this for both Math and Reading scores
num_passing_math = passing_math.groupby("School ID")["math_score"].count()
percent_passing_math = num_passing_math/school_data["size"]

num_passing_reading= passing_reading.groupby("School ID")["reading_score"].count()
percent_passing_reading = num_passing_reading/school_data["size"]


# add newly created "percent passing" columns to our dataframe
school_data["% Passing Math"] = percent_passing_math
school_data["% Passing Reading"] = percent_passing_reading

# calculate overall passing rate, create new column and add to our dataframe
overall_passing_rate = (percent_passing_math + percent_passing_reading)/2
school_data["Overall Passing Rate"] = overall_passing_rate

school_data.set_index("school_name", inplace = True) # set school name as the index

school_data


```

## Math Scores by Grade

* Create a table that lists the average Math Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# create new dataframe containing only information for the specified grade
# regroup by school and calculate average

#9th grade 

nineth_grade = school_data_all.loc[school_data_all["grade"] == "9th"]
nineth_grade_math = nineth_grade.groupby("school_name")

nine_m = nineth_grade_math["math_score"].mean().to_frame()


#10th grade

tenth_grade = school_data_all.loc[school_data_all["grade"] == "10th"]
tenth_grade_math = tenth_grade.groupby("school_name")

ten_m = tenth_grade_math["math_score"].mean().to_frame()


#11th grade

eleventh_grade = school_data_all.loc[school_data_all["grade"] == "11th"]
eleventh_grade_math = eleventh_grade.groupby("school_name")

eleven_m = eleventh_grade_math["math_score"].mean().to_frame()


#12th grade

twelth_grade = school_data_all.loc[school_data_all["grade"] == "12th"]
twelth_grade_math = twelth_grade.groupby("school_name")

twelth_m = twelth_grade_math["math_score"].mean().to_frame()
```

```python
# Merge 
math_merge_1 = nine_m.merge(ten_m, how = "left", on = "school_name", suffixes = (" 9th", " 10th"))

math_merge_2 = eleven_m.merge(twelth_m, how = "left", on = "school_name", suffixes = (" 11th", " 12th"))

math_merge_all = math_merge_1.merge(math_merge_2, how = "left", on = "school_name")


# rename columns
math_by_grade = math_merge_all.rename(columns = {"math_score 9th":"9th", 
                                               "math_score 10th":"10th", 
                                               "math_score 11th":"11th", 
                                               "math_score 12th":"12th"})

math_by_grade
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
school_spending = school_data[["Per Student Budget",
                              "Average Math Score", "Average Reading Score",
                              "% Passing Math", "% Passing Reading", "Overall Passing Rate"]]

budget_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

school_spending["Spending Range (Per Student)"] = pd.cut(school_spending["Per Student Budget"],
                                                        budget_bins, labels = group_names)

binned_spending = school_spending.groupby("Spending Range (Per Student)")

binned_spending["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "Overall Passing Rate"].mean()

```
