```python
import pandas as pd
import numpy as np
import statistics
```

```python
# find files
schools = "Resources/schools_complete.csv"
students = "Resources/students_complete.csv"

#read files
school_df = pd.read_csv(schools)
student_df = pd.read_csv(students)

school_data_all = pd.merge(student_df, school_df, how = 'left', on = 'school_name')

school_data_all.head()
```

```python
school_data_all["school_name"].value_counts() #total students at each school
```

```python
#total number of schools and students

num_of_schools = len(school_data_all["school_name"].unique())
num_of_students = school_data_all["student_name"].value_counts().sum()
```

```python
#total budget
total_budget = school_df["budget"].sum()
```

```python
#average math and reading scores

avg_math_all = student_df["math_score"].mean()
avg_read_all = student_df["reading_score"].mean()

```

```python
passing_math = student_df.loc[student_df["math_score"] >= 70]
passing_read = student_df.loc[student_df["reading_score"] >= 70]
#passing_overall = student_df.loc[(student_df["math_score"] >= 70) | (student_df["reading_score"] >= 70)]



percent_passing_math = passing_math["student_name"].count()/num_of_students
percent_passing_read = passing_read["student_name"].count()/num_of_students
#percent_passing_overall = passing_overall["student_name"].count()/num_of_students
percent_passing_overall = (percent_passing_math + percent_passing_read)/2

```

```python
district_data = pd.DataFrame(data = {"Total Schools" : [num_of_schools], "Total Students" : [num_of_students], 
                                     "Total Budget" : [total_budget], "Average Math Score" :[avg_math_all], 
                                     "Average Reading Score" : [avg_read_all], "% Passing Math" : [percent_passing_math], 
                                     "% Passing Readng" :[percent_passing_read], " Overall Passing Rate" :[percent_passing_overall]})
district_data
```
